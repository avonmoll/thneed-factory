# Detecting Beats

```python
import pyaudio
p = pyaudio.PyAudio()
mic = p.get_default_input_device_info(); print(mic)
s = pyaudio.Stream(p, rate=int(mic['defaultSampleRate']), channels=1,\
                   format=pyaudio.paInt32, input=True, start=True)
s_out = pyaudio.Stream(p, rate=int(mic['defaultSampleRate']), channels=1,\
                   format=pyaudio.paInt32, output=True, start=True)

CHUNKS = 10000
samples = []
signal = []
for i in xrange(10000):
    samples.append(s.read(8))
    signal.append(numpy.frombuffer(samples[i],numpy.int32))
signal = numpy.ravel(signal)
    
for samp in samples:
    s_out.write(samp)
    
s.stop_stream(); s.close()
s_out.stop_stream(); s_out.close()
p.terminate()
```

```python
time = arange(len(signal))/mic['defaultSampleRate']
plot(time,abs(signal/float(0xffffffff)))
title('Raw Signal')
sig2 = [max(abs(signal[i:i+64]/float(0xffffffff))) for i in xrange(0,len(signal),64)]
#print sig2
time2 = [time[i] for i in xrange(0,len(time),64)]
figure()
plot(time2,sig2)
title('%.4f s Moving Maximum'%(64/mic['defaultSampleRate']))

def detectBeats(signal=signal):
    beats = numpy.zeros((len(signal),))
    i = 0
    thresh = 0.080*float(0xffffffff)
    while i < len(signal):
        if signal[i] > thresh:
            beats[i] = 1
            i = i + 8*64
        else:
            i = i + 1
    return beats

myBeats = detectBeats()
```

```python
figure()
plot(time,abs(signal))
width = 64/mic['defaultSampleRate']
bar(time[myBeats>0]-5*width,myBeats[myBeats>0]*float(0xffffffff)/4,width=10*width, color='r')
print len(myBeats[myBeats > 0]), '\n', time[myBeats>0]
```

```python
deltas = numpy.array([time[myBeats>0][i] - time[myBeats>0][i-1] for i in xrange(1,len(time[myBeats>0]))])
print('BPM =~ %d'%(60/deltas.mean()) + '\n' + \
      'var = +/- %.3f'%(60/(deltas.mean()-deltas.var()) - 60/deltas.mean()))
```

## Beat Detection Algorithm (http://stackoverflow.com/questions/12344951/detect-beat-and-play-wav-file-in-a-syncronised-manner)

```python
class SimpleBeatDetection:
    """
    Simple beat detection algorithm from
    http://archive.gamedev.net/archive/reference/programming/features/beatdetection/index.html
    """
    def __init__(self, history = 43):
        self.local_energy = numpy.zeros(history) # a simple ring buffer
        self.local_energy_index = 0 # the index of the oldest element

    def detect_beat(self, signal):

        samples = signal.astype(numpy.int) # make room for squares
        #optimized sum of squares, i.e faster version of (samples**2).sum()
        samples = samples / float(0xffffffff)
        instant_energy = numpy.dot(samples, samples) # normalize

        local_energy_average = self.local_energy.mean()
        local_energy_variance = self.local_energy.var()

        beat_sensibility = (-0.0025714 * local_energy_variance) + 1.15142857
        #beat_sensibility = (-0.0025714 * local_energy_variance) + 
        #print instant_energy, beat_sensibility*local_energy_average
        beat = instant_energy > beat_sensibility * local_energy_average

        self.local_energy[self.local_energy_index] = instant_energy
        self.local_energy_index -= 1
        if self.local_energy_index < 0:
            self.local_energy_index = len(self.local_energy) - 1

        return beat

det = SimpleBeatDetection()
beats = [det.detect_beat(signal[i:i+64])for i in xrange(0,len(signal),64)]
plot(time,signal)
b = bar(time2[beats>0],beats[beats>0]*max(signal),width=64/mic['defaultSampleRate'])
```
